```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
df_mod = pd.read_csv(r"C:\Users\yudhia\Documents\GitHub\FuelPrediction\Dataframe\MOD\MODData_Pandas16102025A.csv")
df_cmd = pd.read_csv(r"C:\Users\yudhia\Documents\GitHub\FuelPrediction\Dataframe\CMDB\BGLData_Pandas16102025B.csv")
```

```python
TARGET = "Fuel"
CANDIDATE_FEATURES = ["OB (Bcm)", "Coal (Ton)", "Distance OB (KM)", "Distance Coal (KM)"]
```

```python
# Keep only columns present in both datasets (features + target)
common_cols = list(set(CANDIDATE_FEATURES + [TARGET]) 
                   & set(df_mod.columns) 
                   & set(df_cmd.columns))

df_mod_aligned = df_mod[common_cols].copy()
df_cmd_aligned = df_cmd[common_cols].copy()

# Ensure numeric types
for c in common_cols:
    df_mod_aligned[c] = pd.to_numeric(df_mod_aligned[c], errors="coerce")
    df_cmd_aligned[c] = pd.to_numeric(df_cmd_aligned[c], errors="coerce")

# Drop rows with missing target or features
df_mod_aligned = df_mod_aligned.dropna(subset=common_cols)
df_cmd_aligned = df_cmd_aligned.dropna(subset=common_cols)

FEATURES = [c for c in common_cols if c != TARGET]
print("Common columns:", common_cols)
print("Features used:", FEATURES)
```

```python
def quick_healthcheck(name, df):
    print(f"\n=== {name}: shape {df.shape} ===")
    print("Missing per column:\n", df.isna().sum())
    print("\nDescribe (numerical):\n", df.describe().T[[ "mean","std","min","25%","50%","75%","max"]])

quick_healthcheck("MOD (Sangatta)", df_mod_aligned)
quick_healthcheck("CMD (Bengalon)", df_cmd_aligned)
```

```python
def outlier_ratio(df, cols):
    ratios = {}
    for c in cols:
        q1, q3 = df[c].quantile([0.25, 0.75])
        iqr = q3 - q1
        lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
        ratios[c] = ((df[c] < lower) | (df[c] > upper)).mean()
    return pd.Series(ratios, name="outlier_ratio")

print("\nOutlier ratio MOD:\n", outlier_ratio(df_mod_aligned, FEATURES+[TARGET]))
print("\nOutlier ratio CMD:\n", outlier_ratio(df_cmd_aligned, FEATURES+[TARGET]))
```

```python
def corr_to_target(df, target):
    c = df.corr(numeric_only=True)[target].drop(target).sort_values(ascending=False)
    return c.to_frame(name="corr_with_"+target)

corr_mod = corr_to_target(df_mod_aligned, TARGET)
corr_cmd = corr_to_target(df_cmd_aligned, TARGET)

print("\nCorrelation with Fuel (MOD):\n", corr_mod)
print("\nCorrelation with Fuel (CMD):\n", corr_cmd)

# Combine for comparison
corr_compare = pd.concat([corr_mod.rename(columns={corr_mod.columns[0]: "MOD"}),
                          corr_cmd.rename(columns={corr_cmd.columns[0]: "CMD"})],
                         axis=1)
print("\nCorrelation compare:\n", corr_compare)
```

```python
# Histograms for each feature & target (per site)
for col in FEATURES + [TARGET]:
    fig, ax = plt.subplots()
    sns.kdeplot(df_mod_aligned[col], label="MOD", fill=True, alpha=0.4)
    sns.kdeplot(df_cmd_aligned[col], label="CMD", fill=True, alpha=0.4)
    ax.set_title(f"Distribution: {col}")
    ax.legend()
    plt.show()

# Fuel vs key drivers (scatter)
for col in FEATURES:
    fig, ax = plt.subplots()
    sns.scatterplot(x=df_mod_aligned[col], y=df_mod_aligned[TARGET], label="MOD")
    sns.scatterplot(x=df_cmd_aligned[col], y=df_cmd_aligned[TARGET], label="CMD")
    ax.set_title(f"Fuel vs {col}")
    ax.set_xlabel(col); ax.set_ylabel(TARGET)
    ax.legend()
    plt.show()
```

```python
def tt_split(df, features, target, test_size=0.2, seed=42):
    X = df[features]
    y = df[target]
    return train_test_split(X, y, test_size=test_size, random_state=seed)

Xtr_mod, Xte_mod, ytr_mod, yte_mod = tt_split(df_mod_aligned, FEATURES, TARGET)
Xtr_cmd, Xte_cmd, ytr_cmd, yte_cmd = tt_split(df_cmd_aligned, FEATURES, TARGET)
```

```python
ridge_pipe = Pipeline([
    ("poly", PolynomialFeatures(degree=2, include_bias=False)),
    ("scaler", StandardScaler()),
    ("ridge", Ridge(alpha=0.1))  # adjust/tune later if needed
])

gbr_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("gbr", GradientBoostingRegressor(random_state=42))
])

models = {
    "Ridge(Poly=2, alpha=0.1)": ridge_pipe,
    "GradientBoosting": gbr_pipe,
}
```

```python
def evaluate_model(name, pipe, Xtr, Xte, ytr, yte):
    # Fit
    pipe.fit(Xtr, ytr)
    # Predict
    ytr_pred = pipe.predict(Xtr)
    yte_pred = pipe.predict(Xte)
    # Metrics
    res = {
        "model": name,
        "train_mse": mean_squared_error(ytr, ytr_pred),
        "test_mse": mean_squared_error(yte, yte_pred),
        "train_mae": mean_absolute_error(ytr, ytr_pred),
        "test_mae": mean_absolute_error(yte, yte_pred),
        "train_r2": r2_score(ytr, ytr_pred),
        "test_r2": r2_score(yte, yte_pred),
    }
    # CV (5-fold, shuffled)
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = cross_val_score(pipe, pd.concat([Xtr, Xte]), pd.concat([ytr, yte]),
                                cv=kf, scoring="r2")
    res.update({
        "cv_r2_mean": cv_scores.mean(),
        "cv_r2_std": cv_scores.std(),
        "cv_r2_each": np.round(cv_scores, 3)
    })
    return res

def benchmark_all(df_name, Xtr, Xte, ytr, yte, models):
    rows = []
    print(f"\n===== Benchmark: {df_name} =====")
    for name, pipe in models.items():
        res = evaluate_model(name, pipe, Xtr, Xte, ytr, yte)
        rows.append(res)
        print(f"\n{name}")
        print(f"Train MSE: {res['train_mse']:.3f} | Test MSE: {res['test_mse']:.3f}")
        print(f"Train MAE: {res['train_mae']:.3f} | Test MAE: {res['test_mae']:.3f}")
        print(f"Train R:  {res['train_r2']:.3f} | Test R:  {res['test_r2']:.3f}")
        print(f"CV R mean: {res['cv_r2_mean']:.3f} | std: {res['cv_r2_std']:.3f} | folds: {res['cv_r2_each']}")
    return pd.DataFrame(rows)

bench_mod = benchmark_all("MOD (Sangatta)", Xtr_mod, Xte_mod, ytr_mod, yte_mod, models)
bench_cmd = benchmark_all("CMD (Bengalon)", Xtr_cmd, Xte_cmd, ytr_cmd, yte_cmd, models)
```

```python
def tidy(df, site_label):
    out = df.copy()
    out.insert(0, "site", site_label)
    # Round for readability
    for c in out.columns:
        if c not in ["model", "site", "cv_r2_each"]:
            out[c] = out[c].astype(float).round(3)
    return out

compare_perf = pd.concat([
    tidy(bench_mod, "MOD"),
    tidy(bench_cmd, "CMD"),
], axis=0)

print("\n=== Side-by-side performance ===\n")
print(compare_perf.drop(columns=["cv_r2_each"]))
```

```python
def plot_pred_actual(pipe, Xte, yte, title):
    ypred = pipe.predict(Xte)
    plt.figure()
    sns.scatterplot(x=yte, y=ypred)
    plt.plot([yte.min(), yte.max()], [yte.min(), yte.max()], "--")
    plt.xlabel("Actual Fuel"); plt.ylabel("Predicted Fuel")
    plt.title(f"Actual vs Predicted: {title}")
    plt.show()

def plot_residuals(pipe, Xte, yte, title):
    ypred = pipe.predict(Xte)
    residuals = yte - ypred
    plt.figure()
    sns.scatterplot(x=ypred, y=residuals)
    plt.axhline(0, linestyle="--")
    plt.xlabel("Predicted Fuel"); plt.ylabel("Residuals (Actual - Pred)")
    plt.title(f"Residuals: {title}")
    plt.show()

# Choose the best model per site by test R
best_mod_name = bench_mod.sort_values("test_r2", ascending=False).iloc[0]["model"]
best_cmd_name = bench_cmd.sort_values("test_r2", ascending=False).iloc[0]["model"]

best_mod = models[best_mod_name].fit(Xtr_mod, ytr_mod)
best_cmd = models[best_cmd_name].fit(Xtr_cmd, ytr_cmd)

plot_pred_actual(best_mod, Xte_mod, yte_mod, f"MOD  {best_mod_name}")
plot_residuals(best_mod, Xte_mod, yte_mod, f"MOD  {best_mod_name}")

plot_pred_actual(best_cmd, Xte_cmd, yte_cmd, f"CMD  {best_cmd_name}")
plot_residuals(best_cmd, Xte_cmd, yte_cmd, f"CMD  {best_cmd_name}")
```

```python
def show_importance_or_coef(name, pipe, feat_names):
    print("\n---", name, "---")
    # Try tree
    try:
        est = pipe.named_steps.get("gbr", None)
        if est is None:
            est = pipe.named_steps.get("rf", None)
        if est is not None and hasattr(est, "feature_importances_"):
            imp = pd.Series(est.feature_importances_, index=feat_names).sort_values(ascending=False)
            print("Feature importances:\n", imp)
            return
    except Exception:
        pass
    # Try linear
    try:
        est = pipe.named_steps.get("ridge", None)
        poly = pipe.named_steps.get("poly", None)
        if est is not None:
            if poly is not None:
                # Expand feature names for poly terms
                poly_names = poly.get_feature_names_out(feat_names)
                coefs = pd.Series(est.coef_, index=poly_names).sort_values(ascending=False)
            else:
                coefs = pd.Series(est.coef_, index=feat_names).sort_values(ascending=False)
            print("Top coefficients:\n", coefs.head(15))
            return
    except Exception:
        pass
    print("No importances/coefficients available.")

show_importance_or_coef(f"MOD  {best_mod_name}", best_mod, FEATURES)
show_importance_or_coef(f"CMD  {best_cmd_name}", best_cmd, FEATURES)
```

```python
def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / np.clip(np.abs(y_true), 1e-9, None))) * 100

print("\nMAPE MOD:", mape(yte_mod, best_mod.predict(Xte_mod)).round(2), "%")
print("MAPE CMD:", mape(yte_cmd, best_cmd.predict(Xte_cmd)).round(2), "%")
```
